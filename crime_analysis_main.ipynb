## SECTION 4: DATA CLEANING


```python
# SECTION 4: DATA CLEANING

# Create a copy of the dataframe for cleaning
df_clean = df.copy()

print(f"Starting shape: {df_clean.shape}")
print(f"   Rows: {df_clean.shape[0]:,}")
print(f"   Columns: {df_clean.shape[1]}")

# 4.1 Standardize Column Names

print("--- 4.1 Standardizing Column Names ---")

print("Current columns:")
print(df_clean.columns.tolist())

# Rename columns to more standard/readable names
column_mapping = {
    'DR_NO': 'DR_Number',
    'Date Rptd': 'Date_Reported',
    'DATE OCC': 'Date_Occurred',
    'TIME OCC': 'Time_Occurred',
    'AREA': 'Area_Code',
    'AREA NAME': 'Area_Name',
    'Rpt Dist No': 'Report_District_Number',
    'Part 1-2': 'Part_1_2',
    'Crm Cd': 'Crime_Code',
    'Crm Cd Desc': 'Crime_Description',
    'Mocodes': 'MO_Codes',
    'Vict Age': 'Victim_Age',
    'Vict Sex': 'Victim_Sex',
    'Vict Descent': 'Victim_Descent',
    'Premis Cd': 'Premise_Code',
    'Premis Desc': 'Premise_Description',
    'Weapon Used Cd': 'Weapon_Code',
    'Weapon Desc': 'Weapon_Description',
    'Status': 'Status_Code',
    'Status Desc': 'Status_Description',
    'Crm Cd 1': 'Crime_Code_1',
    'Crm Cd 2': 'Crime_Code_2',
    'Crm Cd 3': 'Crime_Code_3',
    'Crm Cd 4': 'Crime_Code_4',
    'LOCATION': 'Location',
    'Cross Street': 'Cross_Street',
    'LAT': 'Latitude',
    'LON': 'Longitude'
}

# Rename all columns
df_clean.rename(columns=column_mapping, inplace=True)

print("\nColumn names standardized!")
print("New columns:")
print(df_clean.columns.tolist())

# 4.2 Remove Duplicates

print("--- 4.2 Removing Duplicates ---")

print(f"Before removing duplicates: {df_clean.shape}")
rows_before = df_clean.shape[0]

# Remove duplicate rows
df_clean.drop_duplicates(inplace=True)

rows_after = df_clean.shape[0]
rows_removed = rows_before - rows_after

print(f"After removing duplicates: {df_clean.shape}")
print(f"Removed {rows_removed:,} duplicate rows ({(rows_removed/rows_before*100):.2f}%)")

# 4.3 Convert Data Types

print("--- 4.3 Converting Data Types ---")

print("\nConverting Date_Occurred...")
df_clean['Date_Occurred'] = pd.to_datetime(df_clean['Date_Occurred'], format='%Y %b %d %I:%M:%S %p', errors='coerce')

# Check conversion success
dates_not_converted = df_clean['Date_Occurred'].isna().sum()
dates_converted = df_clean['Date_Occurred'].notna().sum()

print(f"Converted Date_Occurred to datetime")
print(f"   Successfully converted: {dates_converted:,} ({(dates_converted/len(df_clean)*100):.2f}%)")
print(f"   Failed to convert: {dates_not_converted:,} ({(dates_not_converted/len(df_clean)*100):.2f}%)")

if dates_converted > 0:
    print(f"   Date range: {df_clean['Date_Occurred'].min()} to {df_clean['Date_Occurred'].max()}")

# Only remove rows if very few failed (less than 1%)
if dates_not_converted > 0 and (dates_not_converted/len(df_clean)) < 0.01:
    rows_before = len(df_clean)
    df_clean = df_clean.dropna(subset=['Date_Occurred'])
    print(f"   Removed {rows_before - len(df_clean):,} rows with invalid dates")
elif dates_not_converted > 0:
    print(f"   Keeping all rows (some have invalid dates)")

print(f"\nConverting Date_Reported...")
df_clean['Date_Reported'] = pd.to_datetime(df_clean['Date_Reported'], format='%Y %b %d %I:%M:%S %p', errors='coerce')

# Check conversion success
dates_not_converted_rptd = df_clean['Date_Reported'].isna().sum()
dates_converted_rptd = df_clean['Date_Reported'].notna().sum()

print(f"Converted Date_Reported to datetime")
print(f"   Successfully converted: {dates_converted_rptd:,} ({(dates_converted_rptd/len(df_clean)*100):.2f}%)")
print(f"   Failed to convert: {dates_not_converted_rptd:,} ({(dates_not_converted_rptd/len(df_clean)*100):.2f}%)")

# Time_Occurred is already integer (HHMM format like 1430 for 2:30 PM)
print(f"\nTime_Occurred already in correct format (integer HHMM)")

# Ensure numeric columns are correct type
print("\nEnsuring numeric columns are correct type...")
df_clean['Victim_Age'] = pd.to_numeric(df_clean['Victim_Age'], errors='coerce')
df_clean['Latitude'] = pd.to_numeric(df_clean['Latitude'], errors='coerce')
df_clean['Longitude'] = pd.to_numeric(df_clean['Longitude'], errors='coerce')
print(f"Numeric columns verified")

print(f"\nData type conversions complete!")
print(f"   Final shape: {df_clean.shape}")
print(f"   Rows: {df_clean.shape[0]:,}")

# 4.4 Handle Missing Values

print("--- 4.4 Handling Missing Values ---")

# Store original shape
original_shape = df_clean.shape

# Victim_Sex - fill missing with 'X' (Unknown)
missing_sex = df_clean['Victim_Sex'].isna().sum()
df_clean['Victim_Sex'].fillna('X', inplace=True)
print(f"Filled {missing_sex:,} missing Victim_Sex values with 'X' (Unknown)")

# Victim_Descent - fill missing with 'X' (Unknown)
missing_descent = df_clean['Victim_Descent'].isna().sum()
df_clean['Victim_Descent'].fillna('X', inplace=True)
print(f"Filled {missing_descent:,} missing Victim_Descent values with 'X' (Unknown)")

# Weapon_Description - keep as NaN (indicates no weapon)
print(f"Weapon_Description: Keeping {df_clean['Weapon_Description'].isna().sum():,} missing values (indicates no weapon)")

# Premise_Description - fill missing with 'Unknown'
missing_premise = df_clean['Premise_Description'].isna().sum()
df_clean['Premise_Description'].fillna('Unknown', inplace=True)
print(f"Filled {missing_premise:,} missing Premise_Description values with 'Unknown'")

# MO_Codes - keep as is (missing is common)
print(f"MO_Codes: Keeping {df_clean['MO_Codes'].isna().sum():,} missing values (not critical for analysis)")

# Cross_Street - keep as is (missing is common)
print(f"Cross_Street: Keeping {df_clean['Cross_Street'].isna().sum():,} missing values (not critical for analysis)")

print(f"\nMissing value handling complete!")
print(f"   Final shape: {df_clean.shape}")

# 4.5 Handle Outliers in Victim Age

print("--- 4.5 Handling Outliers in Victim Age ---")

print("\nVictim Age Statistics (Before):")
print(df_clean['Victim_Age'].describe())
print(f"Min age: {df_clean['Victim_Age'].min()}")
print(f"Max age: {df_clean['Victim_Age'].max()}")

# Visualize age distribution
plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.hist(df_clean['Victim_Age'], bins=100, edgecolor='black', alpha=0.7)
plt.xlabel('Victim Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Victim Age Distribution (Before Cleaning)', fontsize=14, fontweight='bold')
plt.xlim(-10, 120)
plt.grid(axis='y', alpha=0.3)

# Handle unrealistic ages
# Negative ages and ages > 120 are likely data entry errors
original_count = len(df_clean)

# For negative ages and ages > 120, replace with median
median_age = df_clean[(df_clean['Victim_Age'] >= 0) & (df_clean['Victim_Age'] <= 120)]['Victim_Age'].median()

print(f"\nMedian age (valid range): {median_age}")

# Count how many outliers exist
outliers = ((df_clean['Victim_Age'] < 0) | (df_clean['Victim_Age'] > 120)).sum()
print(f"Number of age outliers found: {outliers:,}")

# Replace outliers with median
df_clean.loc[(df_clean['Victim_Age'] < 0) | (df_clean['Victim_Age'] > 120), 'Victim_Age'] = median_age

print(f"Replaced {outliers:,} outlier ages (< 0 or > 120) with median age: {median_age}")

plt.subplot(1, 2, 2)
plt.hist(df_clean['Victim_Age'], bins=100, edgecolor='black', alpha=0.7, color='green')
plt.xlabel('Victim Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Victim Age Distribution (After Cleaning)', fontsize=14, fontweight='bold')
plt.xlim(-10, 120)
plt.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig(output_path + '02_age_outlier_handling.png', dpi=300, bbox_inches='tight')
plt.show()
print("Age outlier visualization saved")

print("\nVictim Age Statistics (After):")
print(df_clean['Victim_Age'].describe())

# 4.6 Standardize Categorical Values

print("--- 4.6 Standardizing Categorical Values ---")

# Standardize Victim_Sex values
print("\n--- Victim Sex ---")
print("Value counts (before):")
print(df_clean['Victim_Sex'].value_counts())

# Standardize to M, F, X
df_clean['Victim_Sex'] = df_clean['Victim_Sex'].str.strip().str.upper()

# Map variations to standard values
sex_mapping = {
    'M': 'M',
    'F': 'F',
    'X': 'X',
    'H': 'X',  # Sometimes used for unknown
    '-': 'X',
    'N': 'X',  # Non-binary or unknown
}

df_clean['Victim_Sex'] = df_clean['Victim_Sex'].map(sex_mapping).fillna('X')

print("\nValue counts (after):")
print(df_clean['Victim_Sex'].value_counts())
print("Victim_Sex standardized to M, F, X")

# Clean Crime_Description - strip whitespace
print("\n--- Crime Description ---")
df_clean['Crime_Description'] = df_clean['Crime_Description'].str.strip()
print(f"Crime_Description cleaned")
print(f"   Unique crime types: {df_clean['Crime_Description'].nunique()}")
print(f"   Top 5 crimes:")
for idx, (crime, count) in enumerate(df_clean['Crime_Description'].value_counts().head(5).items(), 1):
    print(f"      {idx}. {crime}: {count:,}")

# Clean Area_Name
print("\n--- Area Name ---")
df_clean['Area_Name'] = df_clean['Area_Name'].str.strip()
print(f"Area_Name cleaned")
print(f"   Unique areas: {df_clean['Area_Name'].nunique()}")
print(f"   Areas: {sorted(df_clean['Area_Name'].unique())}")

# Clean Premise_Description
df_clean['Premise_Description'] = df_clean['Premise_Description'].str.strip()
print(f"\nPremise_Description cleaned")

# Clean Status_Description
df_clean['Status_Description'] = df_clean['Status_Description'].str.strip()
print(f"Status_Description cleaned")

print("\nCategorical value standardization complete!")

# 4.7 Create Derived Features for Analysis

print("--- 4.7 Creating Derived Features ---")


# Extract Year
df_clean['Year'] = df_clean['Date_Occurred'].dt.year
print(f"Created 'Year' column")
print(f"   Years in dataset: {sorted(df_clean['Year'].unique())}")
print(f"   Crimes per year:")
for year in sorted(df_clean['Year'].unique()):
    count = (df_clean['Year'] == year).sum()
    print(f"      {year}: {count:,}")

# Extract Month (numeric 1-12)
df_clean['Month'] = df_clean['Date_Occurred'].dt.month
print(f"\nCreated 'Month' column (1-12)")

# Extract Month Name (for better visualization)
df_clean['Month_Name'] = df_clean['Date_Occurred'].dt.month_name()
print(f"Created 'Month_Name' column")

# Extract Day of Week (Monday, Tuesday, etc.)
df_clean['Day_of_Week'] = df_clean['Date_Occurred'].dt.day_name()
print(f"Created 'Day_of_Week' column")

# Extract Day of Week Number (0=Monday, 6=Sunday)
df_clean['Day_of_Week_Num'] = df_clean['Date_Occurred'].dt.dayofweek
print(f"Created 'Day_of_Week_Num' column (0=Monday, 6=Sunday)")

# Create Is_Weekend flag (1=Weekend, 0=Weekday)
df_clean['Is_Weekend'] = df_clean['Day_of_Week_Num'].isin([5, 6]).astype(int)
print(f"Created 'Is_Weekend' column (1=Weekend, 0=Weekday)")
weekend_crimes = (df_clean['Is_Weekend'] == 1).sum()
weekday_crimes = (df_clean['Is_Weekend'] == 0).sum()
print(f"   Weekend crimes: {weekend_crimes:,} ({weekend_crimes/len(df_clean)*100:.1f}%)")
print(f"   Weekday crimes: {weekday_crimes:,} ({weekday_crimes/len(df_clean)*100:.1f}%)")

# Extract Quarter
df_clean['Quarter'] = df_clean['Date_Occurred'].dt.quarter
print(f"\nCreated 'Quarter' column (Q1-Q4)")

# Extract Day of Month
df_clean['Day_of_Month'] = df_clean['Date_Occurred'].dt.day
print(f"Created 'Day_of_Month' column")

# Extract Hour from Time_Occurred (HHMM format -> Hour)
df_clean['Hour'] = (df_clean['Time_Occurred'] // 100).astype(int)

# Fix any invalid hours (>23 or <0)
invalid_hours = ((df_clean['Hour'] < 0) | (df_clean['Hour'] > 23)).sum()
if invalid_hours > 0:
    print(f"   Found {invalid_hours:,} invalid hours, setting to NaN")
    df_clean.loc[(df_clean['Hour'] < 0) | (df_clean['Hour'] > 23), 'Hour'] = np.nan

print(f"Created 'Hour' column (0-23)")

# Create Time_Period categories (Night/Morning/Afternoon/Evening)
def get_time_period(hour):
    if pd.isna(hour):
        return 'Unknown'
    elif 0 <= hour < 6:
        return 'Night'
    elif 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 18:
        return 'Afternoon'
    elif 18 <= hour < 24:
        return 'Evening'
    else:
        return 'Unknown'

df_clean['Time_Period'] = df_clean['Hour'].apply(get_time_period)
print(f"Created 'Time_Period' column (Night/Morning/Afternoon/Evening)")
print(f"   Crime distribution by time period:")
for period in ['Night', 'Morning', 'Afternoon', 'Evening', 'Unknown']:
    count = (df_clean['Time_Period'] == period).sum()
    pct = count / len(df_clean) * 100
    print(f"      {period}: {count:,} ({pct:.1f}%)")

# Create Year-Month column for time series analysis
df_clean['Year_Month'] = df_clean['Date_Occurred'].dt.to_period('M')
print(f"\nCreated 'Year_Month' column for time series analysis")

print("\nAll derived features created successfully!")
print(f"\nTotal columns now: {df_clean.shape[1]}")

# 4.8 Final Data Quality Check

print(f"\nCLEANED DATASET SUMMARY:")
print(f"   Original rows: {df.shape[0]:,}")
print(f"   Final rows: {df_clean.shape[0]:,}")
print(f"   Rows removed: {df.shape[0] - df_clean.shape[0]:,} ({((df.shape[0] - df_clean.shape[0])/df.shape[0]*100):.2f}%)")
print(f"   Original columns: {df.shape[1]}")
print(f"   Final columns: {df_clean.shape[1]}")
print(f"   New columns added: {df_clean.shape[1] - df.shape[1]}")

print(f"\nDATE RANGE:")
print(f"   Start: {df_clean['Date_Occurred'].min().strftime('%B %d, %Y')}")
print(f"   End: {df_clean['Date_Occurred'].max().strftime('%B %d, %Y')}")
print(f"   Duration: {(df_clean['Date_Occurred'].max() - df_clean['Date_Occurred'].min()).days:,} days")
print(f"   Years: {sorted(df_clean['Year'].unique())}")

print(f"\nMISSING VALUES AFTER CLEANING:")
remaining_missing = df_clean.isnull().sum().sum()
print(f"   Total missing values: {remaining_missing:,}")

if remaining_missing > 0:
    missing_summary = df_clean.isnull().sum()
    missing_summary = missing_summary[missing_summary > 0].sort_values(ascending=False)
    print(f"\n   Columns with missing values:")
    for col, count in missing_summary.items():
        pct = (count / len(df_clean)) * 100
        print(f"      {col}: {count:,} ({pct:.2f}%)")

print(f"\nKEY COLUMNS FOR ANALYSIS:")
key_columns = [
    'Date_Occurred', 'Year', 'Month', 'Month_Name', 'Day_of_Week',
    'Hour', 'Time_Period', 'Crime_Description', 'Area_Name',
    'Victim_Age', 'Victim_Sex', 'Latitude', 'Longitude'
]

for col in key_columns:
    if col in df_clean.columns:
        non_null = df_clean[col].notna().sum()
        print(f"   {col}: {non_null:,} non-null values")

# SECTION 5: SUMMARY STATISTICS OF CLEANED DATA

print("\n--- Numerical Columns ---")
display(df_clean.describe())

print("\n--- Key Categorical Columns ---")
categorical_cols = ['Crime_Description', 'Area_Name', 'Victim_Sex',
                   'Day_of_Week', 'Time_Period', 'Status_Description']

for col in categorical_cols:
    if col in df_clean.columns:
        print(f"\n{col} (Top 10):")
        print(df_clean[col].value_counts().head(10))

# SECTION 6: CREATE DATA OVERVIEW DASHBOARD

# Create a comprehensive dashboard
fig = plt.figure(figsize=(20, 12))
gs = fig.add_gridspec(3, 3, hspace=0.3, wspace=0.3)

# 1. Crimes by Year
ax1 = fig.add_subplot(gs[0, :])
crimes_by_year = df_clean['Year'].value_counts().sort_index()
ax1.bar(crimes_by_year.index, crimes_by_year.values, color='steelblue', edgecolor='black', width=0.6)
ax1.set_xlabel('Year', fontsize=12, fontweight='bold')
ax1.set_ylabel('Number of Crimes', fontsize=12, fontweight='bold')
ax1.set_title('Total Crimes by Year (2020-Present)', fontsize=16, fontweight='bold')
ax1.grid(axis='y', alpha=0.3)
# Add value labels on bars
for idx, val in zip(crimes_by_year.index, crimes_by_year.values):
    ax1.text(idx, val, f'{val:,}', ha='center', va='bottom', fontsize=10, fontweight='bold')

# 2. Crimes by Month
ax2 = fig.add_subplot(gs[1, 0])
crimes_by_month = df_clean['Month'].value_counts().sort_index()
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
ax2.bar(crimes_by_month.index, crimes_by_month.values, color='coral', edgecolor='black')
ax2.set_xlabel('Month', fontsize=11, fontweight='bold')
ax2.set_ylabel('Number of Crimes', fontsize=11, fontweight='bold')
ax2.set_title('Crime Distribution by Month', fontsize=13, fontweight='bold')
ax2.set_xticks(range(1, 13))
ax2.set_xticklabels(month_names, rotation=45)
ax2.grid(axis='y', alpha=0.3)

# 3. Crimes by Day of Week
ax3 = fig.add_subplot(gs[1, 1])
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
crimes_by_day = df_clean['Day_of_Week'].value_counts().reindex(day_order)
colors = ['skyblue' if day in ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'] else 'lightcoral' for day in day_order]
ax3.bar(range(7), crimes_by_day.values, color=colors, edgecolor='black')
ax3.set_xlabel('Day of Week', fontsize=11, fontweight='bold')
ax3.set_ylabel('Number of Crimes', fontsize=11, fontweight='bold')
ax3.set_title('Crime Distribution by Day of Week', fontsize=13, fontweight='bold')
ax3.set_xticks(range(7))
ax3.set_xticklabels(['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'], rotation=45)
ax3.grid(axis='y', alpha=0.3)

# 4. Crimes by Time Period
ax4 = fig.add_subplot(gs[1, 2])
time_order = ['Night', 'Morning', 'Afternoon', 'Evening']
crimes_by_time = df_clean['Time_Period'].value_counts().reindex(time_order)
time_colors = ['midnightblue', 'gold', 'orange', 'darkviolet']
ax4.bar(range(4), crimes_by_time.values, color=time_colors, edgecolor='black')
ax4.set_xlabel('Time Period', fontsize=11, fontweight='bold')
ax4.set_ylabel('Number of Crimes', fontsize=11, fontweight='bold')
ax4.set_title('Crime Distribution by Time of Day', fontsize=13, fontweight='bold')
ax4.set_xticks(range(4))
ax4.set_xticklabels(time_order, rotation=45)
ax4.grid(axis='y', alpha=0.3)

# 5. Top 10 Crime Types
ax5 = fig.add_subplot(gs[2, :2])
top_crimes = df_clean['Crime_Description'].value_counts().head(10)
ax5.barh(range(10), top_crimes.values, color='mediumpurple', edgecolor='black')
ax5.set_yticks(range(10))
ax5.set_yticklabels(top_crimes.index, fontsize=10)
ax5.set_xlabel('Number of Crimes', fontsize=11, fontweight='bold')
ax5.set_title('Top 10 Most Common Crime Types', fontsize=13, fontweight='bold')
ax5.invert_yaxis()
ax5.grid(axis='x', alpha=0.3)

# 6. Top 10 Areas
ax6 = fig.add_subplot(gs[2, 2])
top_areas = df_clean['Area_Name'].value_counts().head(10)
ax6.barh(range(10), top_areas.values, color='lightgreen', edgecolor='black')
ax6.set_yticks(range(10))
ax6.set_yticklabels(top_areas.index, fontsize=9)
ax6.set_xlabel('Number of Crimes', fontsize=11, fontweight='bold')
ax6.set_title('Top 10 Areas by Crime Count', fontsize=13, fontweight='bold')
ax6.invert_yaxis()
ax6.grid(axis='x', alpha=0.3)

plt.suptitle('Los Angeles Crime Data Overview Dashboard (2020-Present)',
             fontsize=18, fontweight='bold', y=0.995)

plt.savefig(output_path + '03_data_overview_dashboard.png', dpi=300, bbox_inches='tight')
plt.show()

print("Comprehensive overview dashboard saved!")

# SECTION 7: SAVE CLEANED DATA

# Display the columns in the cleaned dataset
print(f"\nColumns in cleaned dataset ({len(df_clean.columns)} total):")
for i, col in enumerate(df_clean.columns, 1):
    print(f"   {i}. {col}")

# Save the cleaned dataset
df_clean.to_csv(data_path + 'crime_data_cleaned.csv', index=False)

print(f"\nCLEANED DATA SAVED SUCCESSFULLY!")
print(f"Location: {data_path}crime_data_cleaned.csv")
print(f"Shape: {df_clean.shape}")
print(f"Rows: {df_clean.shape[0]:,}")
print(f"Columns: {df_clean.shape[1]}")

print("\nCRITICAL COLUMNS FOR PERSON 2'S EDA:")
eda_columns = {
    'Time Analysis': ['Date_Occurred', 'Year', 'Month', 'Month_Name', 'Day_of_Week', 'Hour', 'Time_Period', 'Year_Month'],
    'Crime Analysis': ['Crime_Description', 'Crime_Code', 'Status_Description'],
    'Location Analysis': ['Area_Name', 'Area_Code', 'Premise_Description', 'Latitude', 'Longitude'],
    'Victim Analysis': ['Victim_Age', 'Victim_Sex', 'Victim_Descent'],
    'Other': ['Weapon_Description', 'Is_Weekend', 'Quarter']
}

for category, columns in eda_columns.items():
    print(f"\n{category}:")
    for col in columns:
        if col in df_clean.columns:
            print(f"   {col}")

# CREATE DATA DICTIONARY


# Create comprehensive data dictionary
data_dictionary = pd.DataFrame({
    'Column_Name': df_clean.columns,
    'Data_Type': [str(dtype) for dtype in df_clean.dtypes],
    'Non_Null_Count': [f"{df_clean[col].notna().sum():,}" for col in df_clean.columns],
    'Null_Count': [f"{df_clean[col].isna().sum():,}" for col in df_clean.columns],
    'Null_Percentage': [f"{(df_clean[col].isna().sum()/len(df_clean)*100):.2f}%" for col in df_clean.columns],
    'Unique_Values': [f"{df_clean[col].nunique():,}" for col in df_clean.columns],
    'Sample_Values': [str(df_clean[col].dropna().head(2).tolist())[:80] + '...' if len(df_clean[col].dropna()) > 0 else 'N/A' for col in df_clean.columns]
})

# Add descriptions
descriptions = {
    'DR_Number': 'Division of Records Number (unique identifier)',
    'Date_Reported': 'Date the crime was reported to police',
    'Date_Occurred': 'Date the crime actually occurred',
    'Time_Occurred': 'Time the crime occurred (HHMM format)',
    'Area_Code': 'LAPD Area code',
    'Area_Name': 'LAPD Area name/division',
    'Report_District_Number': 'Reporting district number',
    'Part_1_2': 'Crime classification (Part 1 or Part 2)',
    'Crime_Code': 'Crime code number',
    'Crime_Description': 'Description of the crime committed',
    'MO_Codes': 'Modus operandi codes',
    'Victim_Age': 'Age of the victim',
    'Victim_Sex': 'Sex of the victim (M/F/X)',
    'Victim_Descent': 'Descent code of the victim',
    'Premise_Code': 'Code for the premise where crime occurred',
    'Premise_Description': 'Description of the premise',
    'Weapon_Code': 'Weapon code used',
    'Weapon_Description': 'Description of weapon used',
    'Status_Code': 'Status code of the case',
    'Status_Description': 'Status description (e.g., Invest Cont)',
    'Crime_Code_1': 'Primary crime code',
    'Crime_Code_2': 'Secondary crime code (if applicable)',
    'Crime_Code_3': 'Tertiary crime code (if applicable)',
    'Crime_Code_4': 'Quaternary crime code (if applicable)',
    'Location': 'Street address of crime location',
    'Cross_Street': 'Cross street nearest to crime',
    'Latitude': 'Latitude coordinate',
    'Longitude': 'Longitude coordinate',
    'Year': '**DERIVED** Year extracted from Date_Occurred',
    'Month': '**DERIVED** Month number (1-12)',
    'Month_Name': '**DERIVED** Month name (January-December)',
    'Day_of_Week': '**DERIVED** Day name (Monday-Sunday)',
    'Day_of_Week_Num': '**DERIVED** Day number (0=Monday, 6=Sunday)',
    'Is_Weekend': '**DERIVED** Weekend flag (1=Weekend, 0=Weekday)',
    'Quarter': '**DERIVED** Quarter (Q1-Q4)',
    'Day_of_Month': '**DERIVED** Day of month (1-31)',
    'Hour': '**DERIVED** Hour of day (0-23)',
    'Time_Period': '**DERIVED** Time period (Night/Morning/Afternoon/Evening)',
    'Year_Month': '**DERIVED** Year-Month for time series analysis'
}

data_dictionary['Description'] = data_dictionary['Column_Name'].map(descriptions).fillna('No description available')

print("\nDATA DICTIONARY:")
display(data_dictionary.head(20))

# Save data dictionary
data_dictionary.to_csv(data_path + 'data_dictionary.csv', index=False)
print("\nData dictionary saved to: data/data_dictionary.csv")

# FINAL SUMMARY REPORT

# Create summary statistics
summary_stats = {
    'Metric': [],
    'Value': []
}

# Dataset statistics
summary_stats['Metric'].extend([
    'Original Dataset Rows',
    'Final Dataset Rows',
    'Rows Removed',
    'Percentage Removed',
    'Original Columns',
    'Final Columns',
    'New Columns Added'
])

summary_stats['Value'].extend([
    f"{df.shape[0]:,}",
    f"{df_clean.shape[0]:,}",
    f"{df.shape[0] - df_clean.shape[0]:,}",
    f"{((df.shape[0] - df_clean.shape[0])/df.shape[0]*100):.2f}%",
    df.shape[1],
    df_clean.shape[1],
    df_clean.shape[1] - df.shape[1]
])

# Date coverage
summary_stats['Metric'].extend([
    'Date Range Start',
    'Date Range End',
    'Days Covered',
    'Years in Dataset'
])

summary_stats['Value'].extend([
    df_clean['Date_Occurred'].min().strftime('%B %d, %Y'),
    df_clean['Date_Occurred'].max().strftime('%B %d, %Y'),
    f"{(df_clean['Date_Occurred'].max() - df_clean['Date_Occurred'].min()).days:,}",
    ', '.join(map(str, sorted(df_clean['Year'].unique())))
])

# Crime statistics
summary_stats['Metric'].extend([
    'Total Unique Crime Types',
    'Most Common Crime',
    'Total Crime Areas',
    'Area with Most Crimes'
])

most_common_crime = df_clean['Crime_Description'].value_counts().index[0]
most_common_crime_count = df_clean['Crime_Description'].value_counts().values[0]
most_crime_area = df_clean['Area_Name'].value_counts().index[0]
most_crime_area_count = df_clean['Area_Name'].value_counts().values[0]

summary_stats['Value'].extend([
    f"{df_clean['Crime_Description'].nunique():,}",
    f"{most_common_crime} ({most_common_crime_count:,})",
    f"{df_clean['Area_Name'].nunique()}",
    f"{most_crime_area} ({most_crime_area_count:,})"
])

# Data quality
summary_stats['Metric'].extend([
    'Total Missing Values',
    'Columns with Missing Data',
    'Data Completeness'
])

total_missing = df_clean.isnull().sum().sum()
cols_with_missing = (df_clean.isnull().sum() > 0).sum()
completeness = ((df_clean.notna().sum().sum()) / (df_clean.shape[0] * df_clean.shape[1]) * 100)

summary_stats['Value'].extend([
    f"{total_missing:,}",
    cols_with_missing,
    f"{completeness:.2f}%"
])

# Create DataFrame
summary_df = pd.DataFrame(summary_stats)

print("\nCOMPREHENSIVE SUMMARY:")
display(summary_df)

# Save summary report
summary_df.to_csv(data_path + 'cleaning_summary_report.csv', index=False)
print("\nSummary report saved to: data/cleaning_summary_report.csv")

# FINAL VERIFICATION

# Test if cleaned data can be loaded and used
try:
    print("\nTesting data loading...")
    test_df = pd.read_csv(data_path + 'crime_data_cleaned.csv')
    test_df['Date_Occurred'] = pd.to_datetime(test_df['Date_Occurred'])

    print("Cleaned data can be loaded and used.")
    print(f"   Shape: {test_df.shape}")
    print(f"   Memory usage: ~{test_df.memory_usage(deep=True).sum() / 1024**2:.1f} MB")

    # Verify critical columns exist
    print("\nVerifying critical columns...")
    critical_columns = [
        'Date_Occurred', 'Year', 'Month', 'Day_of_Week', 'Hour',
        'Crime_Description', 'Area_Name', 'Victim_Age', 'Victim_Sex',
        'Latitude', 'Longitude', 'Time_Period', 'Is_Weekend'
    ]

    missing_critical = [col for col in critical_columns if col not in test_df.columns]

    if missing_critical:
        print(f"WARNING: Missing critical columns: {missing_critical}")
    else:
        print("All critical columns present!")

    # Test basic operations Person 2 will need
    print("\nTesting basic operations for Person 2...")

    # Test 1: Group by year
    yearly_crimes = test_df.groupby('Year').size()
    print(f"Yearly grouping works: {len(yearly_crimes)} years")

    # Test 2: Group by crime type
    crime_types = test_df['Crime_Description'].value_counts()
    print(f"Crime type analysis works: {len(crime_types)} unique crimes")

    # Test 3: Filter by time period
    morning_crimes = test_df[test_df['Time_Period'] == 'Morning']
    print(f"Time period filtering works: {len(morning_crimes):,} morning crimes")

    # Test 4: Date range filtering
    crimes_2023 = test_df[test_df['Year'] == 2023]
    print(f"Date filtering works: {len(crimes_2023):,} crimes in 2023")

except Exception as e:
    print(f"\nERROR: Could not load or test cleaned data!")
    print(f"   Error message: {str(e)}")
    print("\n   Please review the data cleaning steps and try again.")
```
